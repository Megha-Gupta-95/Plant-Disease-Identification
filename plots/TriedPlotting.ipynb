```python
plt.style.use('grayscale')
import matplotlib
import numpy as np
```

```python
lines = []
fig, ax = plt.subplots(figsize=(20, 5))

lines += ax.plot(train_loss["Zfnet"],label="Zfnet train_loss",marker = '^',color= '0')
lines += ax.plot(val_loss["Zfnet"],label = "Zfnet val_loss",marker = '^',linestyle='--',color = '0')
lines += ax.plot(train_loss["Alexnet"],label="Alexnet train_loss",marker = 'o',color = '0.4')
lines += ax.plot(val_loss["Alexnet"],label = "Alexnet val_loss",marker = 'o',linestyle='--',color = '0.4')
plt.grid()
ax.legend(lines[0:3:2],['Zfnet','Alexnet'],title = "Train Loss")
plt.xlabel('Epochs', fontsize=20)
plt.ylabel('Loss', fontsize=20)
#plt.xticks(index, diseases, fontsize=5, rotation=90)
plt.title('Training and Validation Loss') 
plt.xticks(np.arange(0, 25, step=3))  # Set label locations.
from matplotlib.legend import Legend
leg = Legend(ax, lines[1:4:2], ['Zfnet', 'Alexnet'],
             loc='center right',title = "Validation Loss")
ax.add_artist(leg);
```
